```python
#관련 라이브러리 불러오기
import pandas as pd #판다스
import matplotlib as mpl #맷플롭립
import matplotlib.pylab as plt #맷플롭립
import seaborn as sns #시본
import numpy as np # 넘파이

file_path='data/titanic.csv' #data 폴더에 .csv 파일 위치해 있음.
df = pd.read_csv(file_path) #read csv() 함수로 데이터프레임 변환
```

## 데이터 클린징

```python
df.count() #데이터 갯수 확인
```

```python
df.isnull().sum() #결측치 확인
```

```python
#결측치 시각화
sns.heatmap(df.isnull(),cbar=False) #시본 시각화 확인
```

```python
sns.heatmap(df.isnull(), cbar=True) #시본 시각화 확인
```

```python
#Cabin 결측 제거
df['Cabin'].value_counts() #Cabin 삭제 전 결측치 내용 확인
```

```python
df = df.dropna(thresh=int(len(df)*0.5), axis=1)#결측치 제거
sns.heatmap(df.isnull(),cbar=False)#결측치 제거 확인
```

```python
#Embarked 대체
df['Embarked'].value_counts() #범주형 개수 확인 (최다빈도 값을 찾기 위해서)
```

```python
df['Embarked'] = df['Embarked'].fillna('S') #최다빈도 'S'로 대체
```

```python
#Age 대체
df['Age'] = df['Age'].fillna(df['Age'].mean()) #Age의 평균값으로 대체
```

```python
sns.heatmap(df.isnull(), cbar=False) #결측치 제거 최종 확인
```
