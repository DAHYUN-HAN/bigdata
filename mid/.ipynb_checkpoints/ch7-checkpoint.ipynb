```python
df6 = df.math.groupby(df5)
df7 = df6.agg(['count','mean','std','min','max'])
df7
```

```python
import pandas as pd

file_path='data/timeseries.csv'
df=pd.read_csv(file_path)

df.info()
```

```python
df['new_Date'] = pd.to_datetime(df['Date'])
print(df.head())
print('\n')
print(df.info())
print('\n')
print(type(df['new_Date'][0]))
```

```python
df.drop('Date', axis = 1, inplace=True)
df.set_index('new_Date', inplace=True)
print(df.head())
print('\n')
print(df.info())
```

```python
dates=['2020-01-01','2020-03-01','2021-09-01']
ts_dates = pd.to_datetime(dates)
print(ts_dates)
print('\n')
```

```python
pr_day = ts_dates.to_period(freq='D')
print(pr_day)
pr_month = ts_dates.to_period(freq='M')
print(pr_month)
pr_year = ts_dates.to_period(freq='A')
print(pr_year)
```

```python
timestamp_df = pd.date_range(start = '2020-01-01',
                             end = None,
                             periods = 6,
                             freq = 'MS',
                             tz = 'Asia/Seoul')
timestamp_df
```

```python
pr_m = pd.period_range(start = '2020-01-01', end=None, periods = 3, freq = 'M')
print(pr_m)
print('\n')
pr_h = pd.period_range(start = '2020-01-01', end=None, periods = 3, freq = 'H')
print(pr_h)
print('\n')
pr_2h = pd.period_range(start = '2020-01-01', end=None, periods = 3, freq = '2H')
print(pr_2h)
print('\n')
```

```python
import pandas as pd

file_path='data/timeseries.csv'

df=pd.read_csv(file_path)
df.info()
```

```python
df['new_Date'] = pd.to_datetime(df['Date'])
print(df.head())
print('\n')

df['Year'] = df['new_Date'].dt.year
df['Month'] = df['new_Date'].dt.month
df['Day'] = df['new_Date'].dt.day

df.head()
```

```python
df.set_index('new_Date',inplace=True)
df.head()
```

```python
print(df.loc['2015-07'])
print('\n')
print(df.loc['2019-06'])
print('\n')
print(df.loc['2018-06-25':'2018-06-20'])
```

```python
today = pd.to_datetime('2020-03-18')
df['time_diff'] = today - df.index
df.head()
```
