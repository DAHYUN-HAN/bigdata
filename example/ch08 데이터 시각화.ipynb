```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pylab as plt
import seaborn as sns
import numpy as np

plt.rc('font', family='Malgun Gothic')
```

```python
iris = sns.load_dataset("iris")
titaninc = sns.load_dataset("titanic")
```

```python
tips = sns.load_dataset("tips")
lights = sns.load_dataset("flights")
```

```python
sns.set_palette("pastel")
sns.stripplot(x="day", y="total_bill", data=tips)
```

막대그래프

```python
#평균치 빠르게 집계해서 시각화하기
sns.barplot(x="sex", y="tip", data=tips)
```

```python
#데이터 분석은 주로 평균을 보지만 때로는 다른 것들을 보고 싶을 수도 있다.
#예를 들면 데이터에 극단치가 많은 경우 중앙값을 확인하거나 데이터가 범주형인 경우 설문 조사 응답과 같이
#각 카테고리가 몇 번 표시되는지 계산할 수 있다.
#시본은 원하는 집계를 매우 쉽고 간편하게 해낼 수 있다.
#estimator를 지정하면 됨

sns.barplot(x="sex", y="tip", estimator=len, data=tips)
```

```python
#여러 열에서 집단별로 그룹화하고 세부 집단 시각화 하기
#hue 파라미터 추가
sns.barplot(x="sex", y="tip", hue="day", data=tips)
```

```python
#표준편차 오차막대
sns.barplot(x="sex", y="tip", hue="day", data=tips, ci="sd")
```

```python
#요일별 팁 카운트
sns.catplot('day', kind='count', data=tips)
```

박스플롯

```python
sns.boxplot(x="sex", y="tip", data=tips)
```

```python
sns.boxplot(data=tips, orient='h')
```

```python
#박스플롯 활용 이상치 제거 분석
sns.boxplot(data=tips)
```

```python
tips.describe()
```

```python
#total_bill 40이하 tip 8 이하 데이터만 추출 - 조건에 맞는 불린 값 추출
#조건추출 명령어를 다시한번 살펴보시기 바람
(tips['total_bill']<40)&(tips["tip"] <8)
```

```python
#데이터 프레임으로 구성하여 확인하고 새로운 변수 newtips 구성
#조건에 해당하는 10행 데이터가 제거 되었음
newtips=tips[(tips['total_bill']<40)&(tips["tip"] <8)]
newtips
```

```python
#이상치 조정 후 시각화 결과 확인
sns.boxplot(data=newtips)
```
